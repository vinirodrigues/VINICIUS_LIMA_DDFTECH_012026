```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
orders = pd.read_csv("../data/processed/orders_clean.csv")
```

```python
orders["order_purchase_timestamp"] = pd.to_datetime(
    orders["order_purchase_timestamp"]
)

orders["order_month"] = orders["order_purchase_timestamp"].dt.to_period("M")
```

```python
status_count = orders["order_status"].value_counts()

status_count.plot(kind="bar")
plt.title("Quantidade de pedidos por status")
plt.xlabel("Status do pedido")
plt.ylabel("Quantidade")
plt.tight_layout()
plt.show()
```

```python
orders_by_month = orders.groupby("order_month").size()

orders_by_month.plot()
plt.title("Pedidos ao longo do tempo")
plt.xlabel("Ms")
plt.ylabel("Quantidade de pedidos")
plt.tight_layout()
plt.show()
```
